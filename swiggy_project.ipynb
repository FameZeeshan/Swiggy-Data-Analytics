###### 3.1 Function: Price Distribution Histogram

This function creates a histogram showing how restaurant prices are distributed across the dataset.

```python
def visualize_price_distribution():
    """
    Creates a histogram showing the distribution of restaurant prices.

    Purpose: To understand the pricing patterns and identify the most common price ranges.

    Visualization: Histogram with 20 bins
    Color: Salmon with green edges for visual appeal
    """
    # Create a figure with specified size
    plt.figure(figsize=(10, 6))

    # Create histogram of price data, removing any null values
    plt.hist(df['Price'].dropna(), bins=20, color='salmon', edgecolor='green', alpha=0.7)

    # Add title and labels
    plt.title('Price Distribution of Restaurants', fontsize=16, fontweight='bold')
    plt.xlabel('Price (INR)', fontsize=12)
    plt.ylabel('Number of Restaurants', fontsize=12)

    # Add grid for better readability
    plt.grid(axis='y', alpha=0.3)

    # Adjust layout to prevent label cutoff
    plt.tight_layout()

    # Display the plot
    plt.show()

    # Calculate and display insights
    print("\n" + "="*60)
    print("PRICE DISTRIBUTION INSIGHTS")
    print("="*60)
    print(f"Minimum Price: â‚¹{df['Price'].min()}")
    print(f"Maximum Price: â‚¹{df['Price'].max()}")
    print(f"Average Price: â‚¹{df['Price'].mean():.2f}")
    print(f"Median Price: â‚¹{df['Price'].median():.2f}")
    print(f"Most restaurants have prices between â‚¹200 and â‚¹600")
    print("="*60)
```

###### 3.2 Function: Average Ratings Distribution

This function visualizes how customer ratings are distributed across all restaurants.

```python
def visualize_ratings_distribution():
    """
    Creates a histogram showing the distribution of average customer ratings.

    Purpose: To understand customer satisfaction levels across restaurants.

    Visualization: Histogram with 20 bins
    Color: Salmon with black edges
    """
    # Create a figure with specified size
    plt.figure(figsize=(10, 6))

    # Create histogram of ratings data
    plt.hist(df['Avg ratings'].dropna(), bins=20, color='salmon', edgecolor='black', alpha=0.7)

    # Add title and labels
    plt.title('Distribution of Average Customer Ratings', fontsize=16, fontweight='bold')
    plt.xlabel('Average Ratings (out of 5)', fontsize=12)
    plt.ylabel('Number of Restaurants', fontsize=12)

    # Add grid for better readability
    plt.grid(axis='y', alpha=0.3)

    # Adjust layout
    plt.tight_layout()

    # Display the plot
    plt.show()

    # Calculate and display insights
    print("\n" + "="*60)
    print("RATINGS DISTRIBUTION INSIGHTS")
    print("="*60)
    print(f"Minimum Rating: {df['Avg ratings'].min()}")
    print(f"Maximum Rating: {df['Avg ratings'].max()}")
    print(f"Average Rating: {df['Avg ratings'].mean():.2f}")
    print(f"Median Rating: {df['Avg ratings'].median():.2f}")
    print(f"Most restaurants have ratings between 4.0 and 4.5")
    print(f"This indicates generally positive customer experiences.")
    print("="*60)
```

###### 3.3 Function: Delivery Time Distribution

This function shows the distribution of expected delivery times across restaurants.

```python
def visualize_delivery_time_distribution():
    """
    Creates a histogram showing the distribution of delivery times.

    Purpose: To understand typical delivery time expectations on Swiggy.

    Visualization: Histogram with 30 bins
    Color: Light green with black edges
    """
    # Create a figure with specified size
    plt.figure(figsize=(10, 6))

    # Create histogram of delivery time data
    plt.hist(df['Delivery time'].dropna(), bins=30, color='lightgreen', edgecolor='black', alpha=0.7)

    # Add title and labels
    plt.title('Distribution of Delivery Times', fontsize=16, fontweight='bold')
    plt.xlabel('Delivery Time (minutes)', fontsize=12)
    plt.ylabel('Number of Restaurants', fontsize=12)

    # Add grid for better readability
    plt.grid(axis='y', alpha=0.3)

    # Adjust layout
    plt.tight_layout()

    # Display the plot
    plt.show()

    # Calculate and display insights
    print("\n" + "="*60)
    print("DELIVERY TIME INSIGHTS")
    print("="*60)
    print(f"Minimum Delivery Time: {df['Delivery time'].min()} minutes")
    print(f"Maximum Delivery Time: {df['Delivery time'].max()} minutes")
    print(f"Average Delivery Time: {df['Delivery time'].mean():.2f} minutes")
    print(f"Median Delivery Time: {df['Delivery time'].median():.2f} minutes")
    print(f"Most deliveries are promised within 30-70 minutes")
    print("="*60)
```

###### 3.4 Function: Top Cities by Restaurant Count

This function displays the top 10 cities with the highest number of restaurants.

```python
def visualize_top_cities():
    """
    Creates a bar chart showing the top 10 cities by restaurant count.

    Purpose: To identify which cities have the most restaurant options on Swiggy.

    Visualization: Vertical bar chart
    Color: Orange bars
    """
    # Count restaurants per city and get top 10
    city_counts = df['City'].value_counts().head(10)

    # Create a figure with specified size
    plt.figure(figsize=(12, 6))

    # Create bar chart
    city_counts.plot(kind='bar', color='orange', edgecolor='black', alpha=0.8)

    # Add title and labels
    plt.title('Top 10 Cities by Number of Restaurants', fontsize=16, fontweight='bold')
    plt.xlabel('City', fontsize=12)
    plt.ylabel('Number of Restaurants', fontsize=12)

    # Rotate x-axis labels for better readability
    plt.xticks(rotation=45, ha='right')

    # Add grid for better readability
    plt.grid(axis='y', alpha=0.3)

    # Adjust layout
    plt.tight_layout()

    # Display the plot
    plt.show()

    # Display insights
    print("\n" + "="*60)
    print("TOP CITIES INSIGHTS")
    print("="*60)
    for i, (city, count) in enumerate(city_counts.items(), 1):
        print(f"{i}. {city}: {count} restaurants")
    print(f"\nMajor metropolitan cities dominate the restaurant market")
    print("="*60)
```

###### 3.5 Function: Top Areas by Restaurant Count

This function shows the top 10 areas (localities) with the most restaurants.

```python
def visualize_top_areas():
    """
    Creates a bar chart showing the top 10 areas by restaurant count.

    Purpose: To identify restaurant hotspots and popular localities.

    Visualization: Vertical bar chart
    Color: Purple bars
    """
    # Count restaurants per area and get top 10
    area_counts = df['Area'].value_counts().head(10)

    # Create a figure with specified size
    plt.figure(figsize=(12, 6))

    # Create bar chart
    area_counts.plot(kind='bar', color='purple', edgecolor='black', alpha=0.8)

    # Add title and labels
    plt.title('Top 10 Areas by Number of Restaurants', fontsize=16, fontweight='bold')
    plt.xlabel('Area', fontsize=12)
    plt.ylabel('Number of Restaurants', fontsize=12)

    # Rotate x-axis labels for better readability
    plt.xticks(rotation=45, ha='right')

    # Add grid for better readability
    plt.grid(axis='y', alpha=0.3)

    # Adjust layout
    plt.tight_layout()

    # Display the plot
    plt.show()

    # Display insights
    print("\n" + "="*60)
    print("TOP AREAS INSIGHTS")
    print("="*60)
    for i, (area, count) in enumerate(area_counts.items(), 1):
        print(f"{i}. {area}: {count} restaurants")
    print(f"\nCertain popular localities are restaurant hotspots")
    print("="*60)
```

###### 3.6 Function: Most Common Food Types

This function analyzes and displays the most popular cuisine types across all restaurants.

```python
def visualize_common_food_types():
    """
    Creates a bar chart showing the top 10 most common food types.

    Purpose: To understand which cuisines are most widely available.

    Note: Since restaurants can offer multiple cuisines, we split and count each cuisine separately.

    Visualization: Vertical bar chart
    Color: Teal bars
    """
    # Split food types by comma, expand into separate rows, and count
    food_types = df['Food type'].str.split(',', expand=True).stack().str.strip().value_counts().head(10)

    # Create a figure with specified size
    plt.figure(figsize=(12, 6))

    # Create bar chart
    food_types.plot(kind='bar', color='teal', edgecolor='black', alpha=0.8)

    # Add title and labels
    plt.title('Top 10 Most Common Food Types', fontsize=16, fontweight='bold')
    plt.xlabel('Food Type', fontsize=12)
    plt.ylabel('Frequency (Number of Restaurants)', fontsize=12)

    # Rotate x-axis labels for better readability
    plt.xticks(rotation=45, ha='right')

    # Add grid for better readability
    plt.grid(axis='y', alpha=0.3)

    # Adjust layout
    plt.tight_layout()

    # Display the plot
    plt.show()

    # Display insights
    print("\n" + "="*60)
    print("FOOD TYPES INSIGHTS")
    print("="*60)
    for i, (food, count) in enumerate(food_types.items(), 1):
        print(f"{i}. {food}: {count} restaurants")
    print(f"\nNorth Indian, Chinese, and Fast Food are the most abundant cuisines")
    print("="*60)
```

###### 3.7 Function: Correlation Matrix of Numeric Variables

This function creates a correlation matrix to show relationships between numeric variables.

```python
def visualize_correlation_matrix():
    """
    Creates a correlation matrix heatmap for numeric columns.

    Purpose: To understand relationships between price, ratings, total ratings, and delivery time.

    Visualization: Heatmap with color gradient
    Color: Blue gradient (darker = stronger correlation)
    """
    # Select numeric columns for correlation analysis
    numeric_cols = ['Price', 'Avg ratings', 'Total ratings', 'Delivery time']
    correlation = df[numeric_cols].corr()

    # Create a figure with specified size
    fig, ax = plt.subplots(figsize=(8, 6))

    # Create heatmap using matshow
    cax = ax.matshow(correlation, cmap='Blues')

    # Add title
    plt.title('Correlation Matrix of Numeric Variables', pad=20, fontsize=16, fontweight='bold')

    # Set axis labels
    ax.set_xticks(np.arange(len(numeric_cols)))
    ax.set_yticks(np.arange(len(numeric_cols)))
    ax.set_xticklabels(numeric_cols, rotation=45, ha='left')
    ax.set_yticklabels(numeric_cols)

    # Add colorbar
    fig.colorbar(cax)

    # Add correlation values as text
    for i in range(len(numeric_cols)):
        for j in range(len(numeric_cols)):
            text = ax.text(j, i, f'{correlation.iloc[i, j]:.2f}',
                          ha="center", va="center", color="black", fontsize=10)

    # Adjust layout
    plt.tight_layout()

    # Display the plot
    plt.show()

    # Display insights
    print("\n" + "="*60)
    print("CORRELATION MATRIX INSIGHTS")
    print("="*60)
    print("Correlation values range from -1 to 1:")
    print("  â€¢ 1 = Perfect positive correlation")
    print("  â€¢ 0 = No correlation")
    print("  â€¢ -1 = Perfect negative correlation")
    print("\nKey Findings:")
    print("  â€¢ Numeric columns show only weak correlations with each other")
    print("  â€¢ Price and ratings are not strongly correlated")
    print("="*60)
```

###### 3.8 Function: Price Distribution by Top Cities (Boxplot)

This function shows price variations across the top 5 cities using boxplots.

```python
def visualize_price_by_city_boxplot():
    """
    Creates boxplots showing price distribution for top 5 cities.

    Purpose: To compare price ranges and variations across different cities.

    Boxplot elements:
    - Box: Shows the interquartile range (25th to 75th percentile)
    - Line in box: Median value
    - Whiskers: Show data range
    - Dots: Outliers

    Visualization: Multiple boxplots
    """
    # Get top 5 cities by restaurant count
    top5_cities = df['City'].value_counts().head(5).index.tolist()
    data_top5 = df[df['City'].isin(top5_cities)]

    # Create a figure with specified size
    plt.figure(figsize=(10, 6))

    # Create boxplot
    data_top5.boxplot(column='Price', by='City', patch_artist=True)

    # Add title and labels
    plt.title('Price Distribution by City (Top 5 Cities)', fontsize=16, fontweight='bold')
    plt.suptitle('')  # Remove default pandas subtitle
    plt.xlabel('City', fontsize=12)
    plt.ylabel('Price (INR)', fontsize=12)

    # Adjust layout
    plt.tight_layout()

    # Display the plot
    plt.show()

    # Display insights
    print("\n" + "="*60)
    print("PRICE BY CITY INSIGHTS")
    print("="*60)
    for city in top5_cities:
        city_data = df[df['City'] == city]['Price']
        print(f"\n{city}:")
        print(f"  â€¢ Median Price: â‚¹{city_data.median():.2f}")
        print(f"  â€¢ Average Price: â‚¹{city_data.mean():.2f}")
        print(f"  â€¢ Price Range: â‚¹{city_data.min()} - â‚¹{city_data.max()}")
    print("\nSome cities have higher price spreads than others")
    print("="*60)
```

###### 3.9 Function: Price vs Average Ratings (Scatter Plot)

This function creates a scatter plot to explore the relationship between price and ratings.

```python
def visualize_price_vs_ratings():
    """
    Creates a scatter plot showing the relationship between price and ratings.

    Purpose: To check if higher-priced restaurants have better ratings.

    Visualization: Scatter plot with transparency
    Color: Royal blue with alpha for overlapping points
    """
    # Create a figure with specified size
    plt.figure(figsize=(10, 6))

    # Create scatter plot
    plt.scatter(df['Price'], df['Avg ratings'], alpha=0.4, color='royalblue', s=30)

    # Add title and labels
    plt.title('Price vs Average Ratings', fontsize=16, fontweight='bold')
    plt.xlabel('Price (INR)', fontsize=12)
    plt.ylabel('Average Ratings (out of 5)', fontsize=12)

    # Add grid for better readability
    plt.grid(True, alpha=0.3)

    # Adjust layout
    plt.tight_layout()

    # Display the plot
    plt.show()

    # Display insights
    print("\n" + "="*60)
    print("PRICE VS RATINGS INSIGHTS")
    print("="*60)
    print("The scatter plot shows:")
    print("  â€¢ No clear linear relationship between price and ratings")
    print("  â€¢ Good ratings exist across all price ranges")
    print("  â€¢ Expensive doesn't always mean better rated")
    print("  â€¢ Customer satisfaction depends on factors beyond price")
    print("="*60)
```

###### 3.10 Function: Average Ratings by Food Type

This function compares average ratings across different cuisine types.

```python
def visualize_ratings_by_food_type():
    """
    Creates a bar chart showing average ratings for top food types.

    Purpose: To identify which cuisines have the highest customer satisfaction.

    Visualization: Vertical bar chart
    Color: Coral bars
    """
    # Split food types and get top 8
    food_counts = df['Food type'].str.split(',', expand=True).stack().str.strip()
    top_food_types = food_counts.value_counts().head(8).index.tolist()

    # Calculate average ratings for each food type
    food_type_ratings = {}
    for ft in top_food_types:
        food_type_ratings[ft] = df[df['Food type'].str.contains(ft, na=False)]['Avg ratings'].mean()

    # Create a figure with specified size
    plt.figure(figsize=(12, 6))

    # Create bar chart
    plt.bar(food_type_ratings.keys(), food_type_ratings.values(), color='coral', edgecolor='black', alpha=0.8)

    # Add title and labels
    plt.title('Average Ratings by Top Food Types', fontsize=16, fontweight='bold')
    plt.xlabel('Food Type', fontsize=12)
    plt.ylabel('Average Rating (out of 5)', fontsize=12)

    # Rotate x-axis labels for better readability
    plt.xticks(rotation=45, ha='right')

    # Add grid for better readability
    plt.grid(axis='y', alpha=0.3)

    # Add horizontal line for overall average
    plt.axhline(y=df['Avg ratings'].mean(), color='red', linestyle='--', linewidth=2, label='Overall Average')
    plt.legend()

    # Adjust layout
    plt.tight_layout()

    # Display the plot
    plt.show()

    # Display insights
    print("\n" + "="*60)
    print("RATINGS BY FOOD TYPE INSIGHTS")
    print("="*60)
    sorted_ratings = sorted(food_type_ratings.items(), key=lambda x: x[1], reverse=True)
    for i, (food, rating) in enumerate(sorted_ratings, 1):
        print(f"{i}. {food}: {rating:.2f} stars")
    print(f"\nOverall Average Rating: {df['Avg ratings'].mean():.2f}")
    print("Some cuisines like Continental and Desserts tend to have higher ratings")
    print("="*60)
```

###### 3.11 Function: Restaurant Distribution by City (Pie Chart)

This function shows the percentage distribution of restaurants across top cities.

```python
def visualize_city_distribution_pie():
    """
    Creates a pie chart showing restaurant distribution across top 6 cities.

    Purpose: To visualize market share of restaurants in different cities.

    Visualization: Pie chart with percentages
    Colors: Automatic color palette
    """
    # Get top 6 cities
    city_counts = df['City'].value_counts().head(6)

    # Create a figure with specified size
    plt.figure(figsize=(10, 8))

    # Create pie chart
    plt.pie(city_counts.values, labels=city_counts.index, autopct='%1.1f%%', 
            startangle=140, textprops={'fontsize': 11})

    # Add title
    plt.title('Distribution of Restaurants by City (Top 6)', fontsize=16, fontweight='bold', pad=20)

    # Adjust layout
    plt.tight_layout()

    # Display the plot
    plt.show()

    # Display insights
    print("\n" + "="*60)
    print("CITY DISTRIBUTION INSIGHTS")
    print("="*60)
    total_restaurants = city_counts.sum()
    for city, count in city_counts.items():
        percentage = (count / df.shape[0]) * 100
        print(f"{city}: {count} restaurants ({percentage:.1f}% of dataset)")
    print(f"\nTop 6 cities account for {(total_restaurants/df.shape[0])*100:.1f}% of all restaurants")
    print("Restaurant market is dominated by few metropolitan cities")
    print("="*60)
```

###### 3.12 Function: Average Price by City

This function compares average restaurant prices across different cities.

```python
def visualize_avg_price_by_city():
    """
    Creates a bar chart showing average prices in top 8 cities.

    Purpose: To compare cost of dining across different cities.

    Visualization: Vertical bar chart sorted by price
    Color: Sea green bars
    """
    # Calculate average price by city and get top 8
    top8 = df.groupby('City')['Price'].mean().sort_values(ascending=False).head(8)

    # Create a figure with specified size
    plt.figure(figsize=(12, 6))

    # Create bar chart
    top8.plot(kind='bar', color='seagreen', edgecolor='black', alpha=0.8)

    # Add title and labels
    plt.title('Average Price of Dishes by City (Top 8 Cities)', fontsize=16, fontweight='bold')
    plt.xlabel('City', fontsize=12)
    plt.ylabel('Average Price (INR)', fontsize=12)

    # Rotate x-axis labels for better readability
    plt.xticks(rotation=45, ha='right')

    # Add grid for better readability
    plt.grid(axis='y', alpha=0.3)

    # Adjust layout
    plt.tight_layout()

    # Display the plot
    plt.show()

    # Display insights
    print("\n" + "="*60)
    print("AVERAGE PRICE BY CITY INSIGHTS")
    print("="*60)
    for i, (city, price) in enumerate(top8.items(), 1):
        print(f"{i}. {city}: â‚¹{price:.2f}")
    print(f"\nOverall Average Price: â‚¹{df['Price'].mean():.2f}")
    print("Some cities have consistently higher average prices")
    print("="*60)
```

##### 3.13 Function: Delivery Time Distribution by Buckets

This function groups delivery times into 10-minute intervals for better analysis.

```python
def visualize_delivery_time_buckets():
    """
    Creates a bar chart showing restaurant count by delivery time buckets.

    Purpose: To understand which delivery time ranges are most common.

    Note: Delivery times are grouped into 10-minute intervals (e.g., 0-10, 10-20, etc.)

    Visualization: Vertical bar chart
    Color: Red bars
    """
    # Create bins for delivery time (10-minute intervals)
    bins = range(0, 120, 10)
    labels = [f'{b}-{b+10}' for b in bins[:-1]]

    # Create a copy of dataframe to avoid modifying original
    df_copy = df.copy()
    df_copy['Delivery Bucket'] = pd.cut(df_copy['Delivery time'], bins=bins, labels=labels, right=False)

    # Count restaurants in each bucket
    delivery_counts = df_copy['Delivery Bucket'].value_counts().sort_index().head(10)

    # Create a figure with specified size
    plt.figure(figsize=(12, 6))

    # Create bar chart
    delivery_counts.plot(kind='bar', color='red', edgecolor='black', alpha=0.7)

    # Add title and labels
    plt.title('Restaurant Count by Delivery Time Bucket', fontsize=16, fontweight='bold')
    plt.xlabel('Delivery Time (minutes)', fontsize=12)
    plt.ylabel('Number of Restaurants', fontsize=12)

    # Rotate x-axis labels for better readability
    plt.xticks(rotation=45, ha='right')

    # Add grid for better readability
    plt.grid(axis='y', alpha=0.3)

    # Adjust layout
    plt.tight_layout()

    # Display the plot
    plt.show()

    # Display insights
    print("\n" + "="*60)
    print("DELIVERY TIME BUCKETS INSIGHTS")
    print("="*60)
    for bucket, count in delivery_counts.items():
        percentage = (count / len(df_copy)) * 100
        print(f"{bucket} minutes: {count} restaurants ({percentage:.1f}%)")
    print(f"\n30-50 minute delivery times are most common on Swiggy")
    print(f"This represents the optimal balance between speed and logistics")
    print("="*60)
```
